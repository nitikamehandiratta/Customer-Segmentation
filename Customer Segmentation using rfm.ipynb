# RFM Modelling

###### Now we apply RFM modelling to do customer segementation , i.e 
###### we will find out recency which tells how recently has the customer made a transaction with us.
###### we will find out frequency which tells how frequent is the customer in ordering/buying some product from us
###### we will find out Monetary which tells how much does the customer spend on purchasing products from us.

## Recency

```python
df = data.groupby(by='CustomerID',
                        as_index=False)['FIRST_ORDER_DATE'].max()
df.columns = ['CustomerID','LATEST_ORDER_DATE']
recent_date = df['LATEST_ORDER_DATE'].max()
df['Recency_in_days'] = df['LATEST_ORDER_DATE'].apply(
    lambda x: (recent_date - x).days)
df.head()
```

## Frequency

```python
# As in dataset we have column = total orders which tells how many orders each customer id have.
#So it is equal to the frequency
```

```python
df['frequency'] = data['TOTAL_ORDERS']
df.head()
```

## Monetary Value

```python
# As in dataset we have column = REVENUE 
#which tells :
#how much total amount each customer has spend by doing purchase with us
```

```python
df['Monetary'] = data['REVENUE']
df.head()
```

# RFM TABLE

```python
df.drop('LATEST_ORDER_DATE', axis = 1 , inplace = True)
```

```python
df.head()
```

# Rank of RFM

```python
df['R_rank'] = df['Recency_in_days'].rank(ascending=False)
df['F_rank'] = df['frequency'].rank(ascending=True)
df['M_rank'] = df['Monetary'].rank(ascending=True)
df.head()
```

## Normalize Rank of RFM

```python
df['R_rank_norm'] = (df['R_rank']/df['R_rank'].max())*100
df['F_rank_norm'] = (df['F_rank']/df['F_rank'].max())*100
df['M_rank_norm'] = (df['F_rank']/df['M_rank'].max())*100
df.head()
```

```python
df.drop(columns=['R_rank', 'F_rank', 'M_rank'], inplace=True)
 
df.head()
```

# RFM SCORE

##### RFM score is calculated based upon recency, frequency, monetary value normalize ranks. Based upon this score we can divide our customers and rating them on a scale of 5.
##### Formula used for calculating rfm score is : 0.15 *Recency score + 0.28 *Frequency score + 0.57 *Monetary score

```python
df['RFM_Score'] = 0.15*df['R_rank_norm']+0.28 * \
    df['F_rank_norm']+0.57*df['M_rank_norm']
df['RFM_Score'] *= 0.05
df = df.round(2)
df[['CustomerID', 'RFM_Score']].head(7)
```

# Customer Segmentation

```python
# Segmenting the customers into different categories based on ther RFM score
# Potential customers can be classified into differnt classes based on their rfm score

# If rfm score >4.5 : then customer is Top Customer
#4.5 > rfm score > 4 : then customer is High Value Customer
#4>rfm score >3 : then customer is Medium value customer
#3>rfm score>1.6 : then customer is Low-value customer
#rfm score<1.6 :then customer is Lost Customer ie, it is not much potential customer.
```

```python
df["Customer_segment"] = np.where(df['RFM_Score'] >
                                      4.5, "Top Customers",
                                      (np.where(
                                        df['RFM_Score'] > 4,
                                        "High value Customer",
                                        (np.where(
    df['RFM_Score'] > 3,
                             "Medium Value Customer",
                             np.where(df['RFM_Score'] > 1.6,
                            'Low Value Customers', 'Lost Customers'))))))
df[['CustomerID','RFM_Score', 'Customer_segment']].head(20)
```
